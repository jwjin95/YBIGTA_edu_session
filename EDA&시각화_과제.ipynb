### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
null_columns = df.isnull().sum().to_frame('missing_num')
null_columns['missing_%'] = df.isnull().sum() / len(df) * 100 

null_columns = null_columns[null_columns.missing_num>0].sort_values('missing_num', ascending=False)
null_columns
```

```python
# Q2
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
sns.set()
plt.figure(figsize = (15,4))
sns.set_palette('pastel')
sns.barplot(x=null_columns.index, y='missing_%', data=null_columns).set_xticklabels(null_columns.index, rotation=45, ha='right')
# null_columns['missing_%'].plot(kind='bar').set_xticklabels(null_c, rotation=45, ha='right')
```

```python
# Q3
plt.figure(figsize = (15,4))
```
