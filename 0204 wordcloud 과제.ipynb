```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
holmes = open("The-Adventures-of-Sherlock-Holmes.txt", "r", encoding="utf-8").read()
holmes
```

```python
holmes_wc = WordCloud(background_color = "white", collocations = False, stopwords = stopwords,
                     width=400, height=300)
holmes_wc.generate(holmes)
```

```python
holmes_wc.words_
```

```python
holmes_wc.to_image()
```

```python
stopwords.add("said")
holmes_wc.generate(holmes)
holmes_wc.to_image()
```
